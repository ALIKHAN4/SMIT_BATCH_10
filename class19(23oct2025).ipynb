```python
# write the function is_anagram
def is_anagram(test, original):
    test = sorted(test.lower())
    original = sorted(original.lower())
    
    return test == original
    
```

```python
def warn_the_sheep(queue):
    wolf_index = queue.index('wolf') - len(queue)  
    
    if wolf_index == -1:
        return 'Pls go away and stop eating my sheep'
    else:
        return  f"Oi! Sheep number {len(queue) - (queue.index('wolf') + 1) }! You are about to be eaten by a wolf!"
```

```python
def add_length(str_):
    result = []
    value = str_.split(' ')
    for val in value:
        result.append(f'{val} {len(val)}')
    return result
```
