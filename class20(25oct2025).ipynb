```python
def digitize(n):
    result = []
    for num in str(n)[::-1]:
        result.append(int(num))
    return result
```

```python
def str_count(strng, letter):
    counter = 0
    
    for chr in strng:
        if chr == letter:
            counter += 1
        
    return counter
```

```python
def change_count(change):
    data = {
        'penny': 0.01,
        'nickel': 0.05,
        'dime': 0.10,
        'quarter': 0.25,
        'dollar': 1.00,
    }
    total = 0
    for word in change.split(' '):
        
        total += data.get(word)
    result = f"${total:.2f}"
    return result
    
```
