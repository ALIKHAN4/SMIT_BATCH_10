# Positional Argument

```python
def add(num1, num2, num3):
    print(f'num1: {num1}, num2: {num2}, num3: {num3}')
    return num1 + num2 + num3
```

```python
add(4, 5, 6)
```

# Keyword Argument

```python
# def add(num1, num2, num3):

add(num3=3,num1=4,num2=5)
```

```python
add(4,num3=3,num1=2)
```

```python
add()
```

```python
def make_album(artist, album, song=None):
    if song:
        result = {artist: album, 'song':song}
    else:
        result = {artist: album}
    return result
```

```python
make_album('imran khan', 'unforgettable')
```

```python
make_album('atif aslam','kuch bi', 4)
```
