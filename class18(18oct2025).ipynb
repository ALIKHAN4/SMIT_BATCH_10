```python
def make_album(artist, album, song=None):
    if song:
        result = {artist: album, 'song':song}
    else:
        result = {artist: album}
    return result
```

```python
def make_album():
    flag = True
    result = {}
    while flag:
        artist = input('Enter Artist name: ')
        album = input('Enter Album name: ')
        if artist == 'q' or album == 'q':
            flag = False
        else:
            result[artist] = album
    return result
```

```python
make_album()
```

```python
def make_pizza(*toppings):
    print(toppings)
```

```python
make_pizza('pepproni', 'onion')
make_pizza('pepproni', 'onion', 'Bacon')
make_pizza('pepproni', 'onion', 'Bacon', 'Prosciutto')
```

```python
def make_pizza(size, *toppings):
    print(f'Making a {size}-inch pizza with the following toppings:')
    for topping in toppings:
        print(f'- {topping}')
```

```python
make_pizza(15,'Pepproni')
```

```python
make_pizza(16,'pepproni', 'onion', 'Bacon', 'Prosciutto')
```

```python
def build_profile(**info):
    print(info)
```

```python
build_profile(fname = 'ali', lname = 'khan')
build_profile(course = 'python', city = 'karachi')
```

```python
def make_sandwiches(*items):
    print(f'This sandwich includes the following items: ')
    for item in items:
        print(f'- {item}')
```

```python
make_sandwiches('vegetables', 'kabab', 'sauces')
make_sandwiches('vegetables', 'kabab', 'sauces', 'bread')
```

```python
def make_car(manuf, model, **optional_features):
    if optional_features:
        print(f'This car is manufactured by {manuf} and its model is {model} and it has the following optional features: ')
        for key, value in optional_features.items():
            print(f'{key} : {value}')
    else:
        print(f'This car is manufactured by {manuf} and its model is {model}')
```

```python
make_car('Toyota', 'Gli', color='Black')
```

```python
make_car('Toyota', 'Gli')
```
